```python
def convert_currency(value):
    value = value.replace('₹', '').replace(',', '').strip()
    if 'Lakh' in value:
        return float(value.replace('Lakh', '')) * 1203
    elif 'Crore' in value:
        return float(value.replace('Crore', '')) * 120300
    else:
        return float(value) * 0.1203

# "resale_price" sütununa uygulanması
df['resale_price'] = df['resale_price'].apply(convert_currency)

# Sonucun görüntülenmesi
df
```
